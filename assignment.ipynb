# Import important libraries

```python
import pandas as pd
```

# Import data

```python
# import data
data = pd.read_csv("books.csv")
# display first 5 records
data.head()
```

# Data cleaning & pre-processing

### Handle null values

```python
# display null values count for each feature
data.isnull().sum()
```

```python
# Drop books with null values
data.dropna(inplace=True)
# Reset index after droping books
data.reset_index(inplace=True)
data
```

### Handle duplicates

```python
# Check for duplicated books count
duplicated_books_count = data.duplicated().sum()
duplicated_books_count
```

# Data analysis

### Select Harry Potter books

```python
# select books that contains the term "harry poter" in ["title","original_title"] features
harry_potter_books = data.loc[  data["original_title"].str.contains("harry potter" , case=False) | 
                                data["title"].str.contains("harry potter" , case=False)]
# Reset index
harry_potter_books.reset_index(inplace=True)
# Drop auto created unwanted ["index","level_0"] features
harry_potter_books.drop(["index", "level_0"], axis="columns",inplace=True)

harry_potter_books
```

### Average rating of harry poter books

```python
avg_rating_of_harry_potter_books = (harry_potter_books["average_rating"]).sum()/harry_potter_books["average_rating"].count()
print(f"Average rating of harry potter books: {avg_rating_of_harry_potter_books}")
```

### Most selling harry potter books

#### display data related to sales

```python
selling_data = harry_potter_books[["books_count","work_ratings_count","work_text_reviews_count"]]
selling_data
```

#### feature extraction

```python
# extract new feature to identify most selling books
harry_potter_books["book_score"] = selling_data["books_count"]+selling_data["work_ratings_count"]+selling_data["work_text_reviews_count"]
# sort books according to extracted feature
harry_potter_books.sort_values("book_score",ascending=False)
```

#### display most selling books of harry potter (in order)

```python
print("Most selling harry potter books:")
for counter,title in enumerate(harry_potter_books["original_title"]):
    print(f"{counter+1}: {title}")
```
